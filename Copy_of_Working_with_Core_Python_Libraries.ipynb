```python
from datetime import datetime, date, timedelta
now = datetime.now()
print("Now",now)
today = date.today()
print("today",today)
dob= date(2000,1,11)
print("DOB",dob)
yesterday = today - timedelta(days=1)
tomorrow= today+timedelta(days=1)
print("yesterday",yesterday)
print("tomorrow",tomorrow)
formatted = now.strftime("%d-%m-%Y %H:%M:%S")
print("Formatted DateTime:", formatted)
```

```python
import random
print("Randomint" , random.randint(1,100))
print("Random float" , random.random())
colors= [ 'red', 'green', 'blue']
print("random colour",random.choice(colors))
random.shuffle(colors)
print("random shuffle",colors)
print("sample", random.sample(range(1,50),5))
```

```python
from collections import Counter
fruits = [ 'apple', 'banana' ,'mango','apple', 'banana', 'mango']
count = Counter(fruits)
print("Fruit Count:",count)
```

```python
from collections import defaultdict
marks = defaultdict(lambda:0)
marks['Math']=90
marks['Science']=80
print("English(not set):",marks['English'])
```

```python
from collections import namedtuple
Point = namedtuple('Point', 'x y')
p1 = Point(10, 20)
print("Point:", p1)
print("x:", p1.x, "y:", p1.y)
```

```python
from collections import deque
dq = deque([1, 2, 3])
dq.append(4)
dq.appendleft(0)
dq.pop()
dq.popleft()
print("Deque after operations:", dq)
```
